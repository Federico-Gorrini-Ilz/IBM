```python
# LSTM & GRU RNN for Stock Market Prediction
# ---

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

import yfinance as yf
import pandas as pd
from datetime import datetime

from datetime import datetime
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout, GRU
from keras.layers import *
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error as MSE, mean_absolute_error as MAE
from keras.optimizers import Adam, SGD

from IPython.display import display


company = 'WMT'

# Define split ratio variables for train/test split
split_train = 0.80
split_test = 1 - split_train

# Model hyperparameters
window = 60          # Time steps in each window
forecast_N = 30      # Number of days to forecast
epochs = 100          # Increased number of epochs
batch_size = 16      # Smaller batch size for more frequent weight updates
units = 80           # Increased number of units in each LSTM/GRU layer
dropout_rate = 0.15  # Dropout rate to prevent overfitting

# EXTRACT LOAD TRANSFORM (ETL)
# ---
def correct_datetime(date_str):
    # Convert the date string from 'dd/mm/yyyy' to datetime object
    date = datetime.strptime(date_str, '%d/%m/%Y')
    # Convert the datetime object to 'yyyy-mm-dd' format
    return date.strftime("%Y-%m-%d")

# Fetch historical data for Walmart (WMT) from Yahoo Finance
stock = yf.Ticker(company)
df = stock.history(period="max")  # You can specify a period like "1y", "5y", etc.

# Reset the index to get the 'Date' column properly
df.reset_index(inplace=True)
# Convert the 'Date' column to datetime, automatically inferring the format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
# Check for any invalid date entries
print(df[df['Date'].isna()])
# Now that the dates are in the proper datetime format, format them to 'yyyy-mm-dd'
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
# Sort the DataFrame by 'Date' and reset the index
df = df.sort_values('Date').reset_index(drop=True)
# Focus on the 'Close' prices, which is typical for stock price analysis
df['Price'] = df['Close'].astype(float)

# --- New Data Start/End Definitions ---
# Explicitly define start and end of dataset
data_start = '2013-08-15'
data_end = '2022-08-15'

# Filter the dataframe based on date range
df_filtered = df[(df['Date'] >= data_start) & (df['Date'] <= data_end)]

# Split the dataset into train and validation based on the range
split_train = 0.80  # You can adjust this ratio as needed
num_shape = int(split_train * df_filtered.shape[0])

train = df_filtered.iloc[:num_shape, 1:2].values  # Price data (excluding Date column)
val = df_filtered.iloc[num_shape:, 1:2].values    # Price data (excluding Date column)

display(df_filtered.shape)
display(df_filtered.head())  # Display first few rows of the train data
display(df_filtered.tail())  # Display first few rows of the validation data


# --- Scaling ---
scaler = StandardScaler()
train_scaled = scaler.fit_transform(train)

X_train = []
y_train = []


for i in range(window, num_shape):
    X_train_ = np.reshape(train_scaled[i - window:i, 0], (window, 1))
    X_train.append(X_train_)
    y_train.append(train_scaled[i, 0])
X_train = np.stack(X_train)
y_train = np.stack(y_train)

# PLOT 1: Stock Price History
plt.figure(figsize=(20, 7))
plt.plot(df_filtered['Date'].values, df_filtered['Price'].values, label='Stock Price', color='red')
plt.xticks(np.arange(100, df_filtered.shape[0], 200))
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price History')
plt.legend()
plt.show()


# LSTM Architecture
LSTM_model = Sequential()

LSTM_model.add(LSTM(units, return_sequences=True, input_shape=(X_train.shape[1], 1)))
LSTM_model.add(Dropout(dropout_rate))

LSTM_model.add(LSTM(units))
LSTM_model.add(Dropout(dropout_rate))

# Adding the output layer
LSTM_model.add(Dense(units=1))
LSTM_model.summary()

# Training
LSTM_model.compile(optimizer='adam', loss='mean_absolute_error')
LSTM_model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)

# Validation inputs
df_volume = np.vstack((train, val))
inputs = df_volume[df_volume.shape[0] - val.shape[0] - window:]
inputs = inputs.reshape(-1, 1)
inputs = scaler.transform(inputs)

num_2 = df_volume.shape[0] - num_shape + window

X_val = []

for i in range(window, num_2):
    X_val_ = np.reshape(inputs[i - window:i, 0], (window, 1))
    X_val.append(X_val_)

X_val = np.stack(X_val)

LSTM_predict = LSTM_model.predict(X_val)
LSTM_predict = scaler.inverse_transform(LSTM_predict)

# MSE and MAE for LSTM prediction
print("LSTM MSE:", MSE(LSTM_predict, val))
print("LSTM MAE:", MAE(LSTM_predict, val))


# PLOT 2: LSTM Stock Price Prediction
plt.figure(figsize=(20, 7))
plt.plot(df_filtered['Date'].values[num_shape:], df_volume[num_shape:], color='red', label='Real Stock Price')
plt.plot(df_filtered['Date'][-LSTM_predict.shape[0]:].values, LSTM_predict, color='grey', linestyle='--', label='Predicted Stock Price')
plt.xticks(np.arange(100, df_filtered[num_shape:].shape[0], 200))
plt.title('LSTM Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()


# LSTM N-day forecast
# ---
LSTM_pred_ = LSTM_predict[-1].copy()  # Take the last prediction from the model
prediction_full = []
window = 60
df_copy = df_filtered.iloc[:, 1:2][1:].values  # Starting from the second data point

for j in range(forecast_N):
    df_ = np.vstack((df_copy, LSTM_pred_))  # Append the previous prediction to the data
    train_ = df_[:num_shape]
    test_ = df_[num_shape:]

    df_volume_ = np.vstack((train_, test_))

    # Prepare inputs for prediction
    inputs_ = df_volume_[df_volume_.shape[0] - test_.shape[0] - window:]
    inputs_ = inputs_.reshape(-1, 1)
    inputs_ = scaler.transform(inputs_)

    X_test_2 = []

    for k in range(window, num_2):
        X_test_3 = np.reshape(inputs_[k - window:k, 0], (window, 1))
        X_test_2.append(X_test_3)

    X_test_ = np.stack(X_test_2)
    LSTM_predict_ = LSTM_model.predict(X_test_, verbose=0)
    LSTM_pred_ = scaler.inverse_transform(LSTM_predict_)
    prediction_full.append(LSTM_pred_[-1][0])  # Append the last prediction
    df_copy = df_[j:]  # Update df_copy to include the new prediction

LSTM_prediction_full_new = np.vstack((LSTM_predict, np.array(prediction_full).reshape(-1, 1)))

# --- Generate next N days of dates ---
df_date = df_filtered[['Date']]

# Generate the next N days of dates starting from the last date in df_date
new_dates = pd.date_range(start=pd.to_datetime(df_date['Date'].iloc[-1]) + pd.DateOffset(days=1), periods=forecast_N, freq='D')
new_dates_df = pd.DataFrame(new_dates.strftime("%Y-%m-%d"), columns=['Date'])

# Concatenate the new dates to df_date
df_date = pd.concat([df_date, new_dates_df], ignore_index=True).reset_index(drop=True)


# PLOT 3: LSTM Stock Price Prediction (30-day forecast)
# Use df_date for the forecast period
plt.figure(figsize=(20, 7))
plt.plot(df_filtered['Date'].values[num_shape:], df_volume[num_shape:], color='red', label='Real Stock Price')
plt.plot(df_date['Date'][-LSTM_prediction_full_new.shape[0]:].values, LSTM_prediction_full_new, color='grey', linestyle='--', label='Predicted Stock Price')
plt.xticks(np.arange(100, df_filtered[num_shape:].shape[0], 200))
plt.title('LSTM Stock Price Prediction (30-day forecast)')
plt.xlabel('Date')
plt.ylabel('Price (US$)')
plt.legend()
plt.tight_layout()
plt.show()


# ---------- Repeat the GRU model code with similar adjustments ----------
# GRU Model
GRU_model = Sequential()

GRU_model.add(GRU(units, return_sequences=True, input_shape=(X_train.shape[1], 1)))
GRU_model.add(Dropout(dropout_rate))

GRU_model.add(GRU(units))
GRU_model.add(Dropout(dropout_rate))

GRU_model.add(Dense(units=1))
GRU_model.summary()

# Training
GRU_model.compile(optimizer='adam', loss='mean_absolute_error')
GRU_model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)

# GRU predictions
GRU_predict = GRU_model.predict(X_val)
GRU_predict = scaler.inverse_transform(GRU_predict)

# MSE and MAE for GRU prediction
print("GRU MSE:", MSE(GRU_predict, val))
print("GRU MAE:", MAE(GRU_predict, val))

# PLOT 4: GRU Stock Price Prediction
plt.figure(figsize=(20, 7))
plt.plot(df_filtered['Date'].values[num_shape:], df_volume[num_shape:], color='red', label='Real Stock Price')
plt.plot(df_filtered['Date'][-GRU_predict.shape[0]:].values, GRU_predict, color='grey', linestyle='--', label='Predicted Stock Price')
plt.xticks(np.arange(100, df_filtered[num_shape:].shape[0], 200))
plt.title('GRU Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()


# GRU N-day forecast
# ---
GRU_pred_ = GRU_predict[-1].copy()
prediction_full = []
window = 60
df_copy = df_filtered.iloc[:, 1:2][1:].values

for j in range(forecast_N):
    df_ = np.vstack((df_copy, GRU_pred_))
    train_ = df_[:num_shape]
    test_ = df_[num_shape:]

    df_volume_ = np.vstack((train_, test_))

    inputs_ = df_volume_[df_volume_.shape[0] - test_.shape[0] - window:]
    inputs_ = inputs_.reshape(-1, 1)
    inputs_ = scaler.transform(inputs_)

    X_test_2 = []

    for k in range(window, num_2):
        X_test_3 = np.reshape(inputs_[k - window:k, 0], (window, 1))
        X_test_2.append(X_test_3)

    X_test_ = np.stack(X_test_2)
    GRU_predict_ = GRU_model.predict(X_test_, verbose=0)
    GRU_pred_ = scaler.inverse_transform(GRU_predict_)
    prediction_full.append(GRU_pred_[-1][0])
    df_copy = df_[j:]

GRU_prediction_full_new = np.vstack((GRU_predict, np.array(prediction_full).reshape(-1, 1)))


# PLOT 5: GRU Stock Price Prediction (30-day forecast)
plt.figure(figsize=(20, 7))
plt.plot(df_filtered['Date'].values[num_shape:], df_volume[num_shape:], color='red', label='Real Stock Price')
plt.plot(df_date['Date'][-GRU_prediction_full_new.shape[0]:].values, GRU_prediction_full_new, color='grey', linestyle='--', label='Predicted Stock Price')
plt.xticks(np.arange(100, df_filtered[num_shape:].shape[0], 200))
plt.title('GRU Stock Price Prediction (30-day forecast)')
plt.xlabel('Date')
plt.ylabel('Price (US$)')
plt.legend()
plt.tight_layout()
plt.show()

# PLOT 6: LSTM vs GRU Price Prediction (N-day forecast)
plt.figure(figsize=(20, 7))
plt.plot(df_date['Date'][-LSTM_prediction_full_new.shape[0]:].values, LSTM_prediction_full_new, color='blue', label='LSTM Prediction')
plt.plot(df_date['Date'][-GRU_prediction_full_new.shape[0]:].values, GRU_prediction_full_new, color='red', label='GRU Prediction')
plt.xticks(np.arange(100, df_date[2000:].shape[0], 200))
plt.title('LSTM vs GRU Price Prediction (30-day forecast)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
